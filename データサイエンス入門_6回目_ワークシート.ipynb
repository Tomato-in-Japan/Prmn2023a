# **CSVファイルの読み込みと内容表示**
　chitose.csvファイルをPandasで読み込みなさい。ファイルはCSV形式のファイルである。本ファイルはヘッダがあるcsvである。読み込んだらdata変数に格納し、data変数を表示しなさい。

表示結果の例は
```
         date  temp_max  temp_min
0    2022/1/1      -7.5     -21.4
1    2022/1/2       1.1     -20.6
2    2022/1/3      -5.2     -15.9
3    2022/1/4      -6.2     -20.2
```

ヒント：
Pandasでcsvファイルを読み込んで変数に格納するには、data = pd.read_csv('chitose.csv')とすればよい。

````python
import pandas as pd
data01 = pd.read_csv('chitose.csv')
print(data01.head())
````

# **課題1**
　読み込んだファイルは、国土交通省の気象庁のサイトが提供する「過去の気象データ・ダウンロード」から取得したデータである。地点は北海道千歳市である。データの項目は「日付」「日最高気温（単位：℃）」「日最低気温（単位：℃）」である。期間は2022年1月1日から2月28日である。英字で扱えるように「日最高気温」を「temp_max」、「日最低気温」を「temp_min」で表現している。
読み込んだデータから最高気温、最低気温、平均気温を求めよ。平均気温は、「日最高気温（単位：℃）」と「日最低気温（単位：℃）」のそれぞれの平均値を求めてください。


````python
print('日最高気温の最高気温',pd.DataFrame.max(data01['temp_max']),'℃')
print('日最高気温の最低気温',pd.DataFrame.min(data01['temp_max']),'℃')
print('日最高気温の平均気温',pd.DataFrame.mean(data01['temp_max']),'℃')
print('日最低気温の最高気温',pd.DataFrame.max(data01['temp_min']),'℃')
print('日最低気温の最低気温',pd.DataFrame.min(data01['temp_min']),'℃')
print('日最低気温の平均気温',pd.DataFrame.mean(data01['temp_min']),'℃')
````

# **課題2**
　次に、降雪量について調べる。降雪量はchitose-snow.csvファイルにデータがある。chitose-snow.csvファイルを読み込んで、課題１のデータと結合して１つの表にしなさい。その表のデータから最も降らなかった日の量、最も降った日の量、降雪量の平均を計算して表示しなさい。
（補足：降雪量の単位はcmです）





````python
data02 = pd.read_csv('chitose-snow.csv')
data03=pd.concat([data01,data02],axis=1)
print(data03.head())
print('最大降雪量',pd.DataFrame.max(data03['snowfall']),'cm')
print('最小降雪量',pd.DataFrame.min(data03['snowfall']),'cm')
print('平均降雪量',pd.DataFrame.mean(data03['snowfall']),'cm')
````

# **課題4**
　降雪量についてさらに調べる。本期間（二ヶ月）において雪が降った日は何日あるか調べなさい。

ヒント：
条件（雪が降った）はqueryメソッドで実現できる。個数を数えるにはcountメソッドを用いれば良い。

````python
df2=data03.query('snowfall>0').count()["snowfall"]
print('2ヶ月で雪が降った日は',df2,'日')
````
